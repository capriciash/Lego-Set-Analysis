```python
import pandas as pd
```

```python
# Import the sets data
sets = pd.read_csv(".\data\sets.csv")
sets.head()
```

```python
# Import the themes data
themes = pd.read_csv(".\data\themes.csv")
themes.head()
```

```python
# Add theme to sets
themes.rename({'id': 'theme_id', 'parent_id': 'parent_theme_id', 'name': 'theme_name'}, axis=1,inplace=True)
sets = sets.merge(themes, how='left', on='theme_id')
sets.head()
```

```python
# Add parent theme to sets
del themes['parent_theme_id']
themes.rename({'theme_id': 'parent_theme_id', 'theme_name': 'parent_theme_name'}, axis=1,inplace=True)

sets = sets.merge(themes, how='left', on='parent_theme_id')
sets.head()
```
